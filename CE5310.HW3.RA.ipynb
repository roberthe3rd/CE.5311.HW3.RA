## Problem 1 
Solve the following integral using numerical methods:
integral from 0 to 1 of : (1+x**2)^0.5
Use a: Trapezoidal rule, b: Simpson's rule, c: Romberg Integration, d: Gauss quadrature. This problem has four parts. They were written so they can be run all together or separately. Note that 1d the Gauss Quadrature requires use of a separate file not currently located within this notebook.

```python
# Import Libraries
import numpy as np

# 1a. Trapezoidal Rule
# Beginning with trapezoidal rule as used by Uddameri within his Romberg Int
# Additional code ideas from Berkley online PDF about numerical methods:
# https://pythonnumericalmethods.berkeley.edu/notebooks/chapter21.03-Trapezoid-Rule.html

# Define order and integration limits
a = 0.0     # Lower limit of integration
b = 1.0     # Upper limit of integration
nTR = 500     # Set number of intervals/trapezoids
hTR = (b-a) / (nTR-1)       # width of intervals
xTR = np.linspace(a,b,nTR)  # returns evenly spaced numbers over the intervals

# Define base function to compute
fTR = np.sqrt(xTR**2+1)

# Trapezoid method can be simplified into the form: 
# integral from a to be f(x)dx = (h/2)*(f(x0)+2*(sum from i=1 to (n-1) of f(xi)) + f(xn))

answerTR = (hTR/2)*(fTR[0] + 2*sum(fTR[1:nTR-1]) + fTR[nTR-1])

print(answerTR)

# 1b. Simpson's rule
# With additional help from berkley
# https://pythonnumericalmethods.berkeley.edu/notebooks/chapter21.04-Simpsons-Rule.html

# Import Libraries
import numpy as np

a = 0.0     # Lower limit of integration
b = 1.0     # Upper limit of integration
nSR = 1000000     # Set number of intervals/trapezoids
hSR = (b-a) / (nSR-1)       # width of intervals
xSR = np.linspace(a,b,nSR)  # returns evenly spaced numbers over the intervals

# Define base function
fSR = np.sqrt(xSR**2+1)

# Simpson's 1/3 rule:
Simple3 = (hSR/3)*(fSR[0] + 2*sum(fSR[:nSR-2:2]) + 4*sum(fSR[1:nSR-1:2]) + fSR[nSR-1])

# Print results. It is noted that this method seems to take significantly
# higher 'n' values to get accurate results
print(Simple3)

# 1c. Romberg Integration
# Romberg Integration by Uddameri from class notes provided
# Import Libraries
import numpy as np

def R(x):
    z = np.sqrt(x**2+1)
    return (z)

# Define order and integration limits

a = 0.0     # Lower limit of integration
b = 1.0     # Upper limit of integration

#      ***** Do not recommend using N (NRI) greater than 20 *****

NRI = 20      # Order of the Romberg integral

# Define the matrix to store values
TRI = np.zeros([NRI,NRI])     # Initialize results matrix
Rxa = R(a)              # compute the integrand at LL
Rxb = R(b)              # compute the integrand at UL
sum1 = Rxa + Rxb        # Sum of UL and LL values
h = (b-a)               # Initial panel width
TRI[0,0] = h/2*(sum1)     # compute single panel estimate

K = np.arange(1,NRI,1)    # Create an array of K

for k in K:
    h = (b-a)/2**k              #update panel width
    idx = np.arange(1,(2**k))   #create summation idx
    z = a + idx*h               # create intermediate values
    fsum = np.sum(R(z))         # sum intermediate value
    TRI[k,0] = h/2*(sum1 + 2*fsum)   # apply trapezoidal rule
    
    for j in np.arange(1,k+1,1):  # Apply richardson extrapolation
        num1 = 1/(4**j-1)
        num2 = 4**j*TRI[(k),(j-1)]
        num3 = TRI[(k-1),(j-1)]
        TRI[k,j] = num1*(num2-num3)   # Obtain refined estimates
       
ResRI = np.diagonal(TRI)    # collect diagonal terms

print(ResRI[NRI-1])           # Print most accurate result of set



# 1d. Gauss Quadrature
# Gauss Quadrature by Uddameri from class notes provided
# Import Libraries
import numpy as np
import pandas as pd
import os

# Function to be integrated
def Qx(x):
    z = np.sqrt(1+x**2)
    return (z)

# Function for coordinate transformation
def newcoord(z,a,b):
    xnew = 0.5*((b-a)*(z) + (a+b))
    return xnew

# set working directory and read weights file
# weights were obtained from using Keisan online calculator

#   *****  OTHER USERS WILL NEET TO UPDATE THE LOCATIONS OF THE WEIGHTS FILE TO OPERATE *****

path = 'C:/Users/rober/Desktop/TTU.Classes/CE.5310.Numerical.Methods\Practice code'
os.chdir(path)

# File with weights and abscissa  .csv created from https://keisan.casio.com/exec/system/1329114617
gq = pd.read_csv('Gauss100.csv')   

xi = gq['xi']   # extract absicssa
wi = gq['wi']   # extract weights

a = 0.0 # user must set lower limit of integration
b = 1.0 # user must set upper limit of integration


zi = newcoord(xi,a,b)   # get new coords
Qzi = Qx(zi)   # compute function with new coords
resGQ = np.sum(Qzi*wi)*(b-a)/2   # apply gauss quad formula
print(resGQ)
```
